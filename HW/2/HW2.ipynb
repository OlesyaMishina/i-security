```python
import pandas as pd
data = pd.read_csv(r'C:\GB\Метавселенные\I_Security\HW\2\Перечень сформированных угроз.csv')
data.sample(15)
```

```python

statistic_measures = pd.DataFrame(columns=['Код меры защиты', 'Название меры защиты', 'Количество баллов'])
statistic_measures
i=0
for row in data['Меры защиты']:
    rows = row.split(';\r\n')
    for d in rows:
            statistic_measures.at[i, 'Код меры защиты'] = d.split()[0].strip()
            statistic_measures.at[i, 'Название меры защиты'] = d.replace(d.split()[0], '')
            i+=1 
statistic_measures
```

```python
i=0
for s in statistic_measures['Код меры защиты']:
#     statistic_measures[statistic_measures['Код меры защиты']==i]['Количество баллов'] = statistic_measures[statistic_measures['Код меры защиты']==i].count()
    statistic_measures.at[i, 'Количество баллов'] = statistic_measures[statistic_measures['Код меры защиты']==s].count()['Код меры защиты']
    i+=1
```

```python
statistic_measures
```

```python
statistic_measures=statistic_measures.drop_duplicates().sort_values(by='Количество баллов', ascending=False).reset_index(drop=True)
statistic_measures.head(50)
```

```python
statistic_measures.to_csv(r'C:\GB\Метавселенные\I_Security\HW\2\Statistic_measures.csv', index=True)
```
